In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# Your code goes here

import pandas as pd

url_1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
url_2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
url_3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'

df_1 = pd.read_csv(url_1)
df_2 = pd.read_csv(url_2)
df_3 = pd.read_csv(url_3)

#Function to clean column names
def rename_columns(df: pd.DataFrame):
    df.copy()
    df.rename(columns={'GENDER':'Gender','ST':'State'},inplace=True)
    df.rename(columns={ col: col.replace(" ","_").lower() for col in df.columns },inplace=True)
    df.columns
    return df

df_1 = rename_columns(df_1)
df_2 = rename_columns(df_2)
df_3 =rename_columns(df_3)


df_1_2 = pd.merge(df_1,df_2[[col for col in df_2.columns if col not in df_1.columns or col == 'customer']], on='customer',how='outer')
df = pd.merge(df_1_2,df_3[[col for col in df_3.columns if col not in df_1_2.columns or col == 'customer']], on='customer',how='outer')
df
```

```python
#clean gender column values
df['gender'] = df['gender'].apply(lambda x: x if x in ['M', 'F'] else None)
#clean state column to a readable format
df['state'] = df['state'].apply(lambda x: x[:2].upper() if x in ['Oregon','California','Arizona','Nevada','Washington','Cali'] else None)
#clean education column
df['education'] = df['education'].replace('Bachelors','Bachelor')
#cleaned customer_life_time_value column
df['customer_lifetime_value'] = df['customer_lifetime_value'].replace('%','',regex=True).astype(float)
#cleaning vehicle_class column format
df['vehicle_class'] = df['vehicle_class'].apply(lambda x: 'Luxury' if x in ['Sports Car', 'Luxury SUV', 'Luxury Car'] else x)
#dropping any n/a columns that have at least 3 null values per row. Data will not be useful.
df = df.dropna(thresh=3)
#formatting customer_life_time_value column
df['customer_lifetime_value'] = df['customer_lifetime_value'].replace('%','',regex=True).astype(float)
#formatting number of open complaints column
df['number_of_open_complaints'] = df['number_of_open_complaints'].apply(lambda x: int(x.split('/')[1]) if isinstance(x, str) and len(x.split('/')) > 1 else 0)
df['number_of_open_complaints'] = df['number_of_open_complaints'].astype(int)
#formatting total_claim_amount column
df['total_claim_amount'] = round(df['total_claim_amount'],2)
df['customer_lifetime_value'] = df['customer_lifetime_value'].apply(lambda x: '{:,}'.format(x) if pd.notnull(x) else None)
df['income'] = df['income'].apply(lambda x: '{:,}'.format(x) if pd.notnull(x) else None)
df['total_claim_amount'] = df['total_claim_amount'].apply(lambda x: '{:,}'.format(x) if pd.notnull(x) else None)
df
```
